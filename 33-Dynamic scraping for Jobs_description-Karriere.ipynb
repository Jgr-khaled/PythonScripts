```python
from selenium import webdriver
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver.common.keys import Keys

from bs4 import BeautifulSoup
import requests


# Create a new instance of the Chrome driver
driver = webdriver.Chrome(executable_path = 'chromedriver'  )

# Navigate to the website you want to interact with
driver.get("https://www.karriere.at/")       

# Create an instance of ActionChains
actions = ActionChains(driver)

#WebDriverWait(driver, 20)

Alle_ablehnen  = ' //*[@id="onetrust-reject-all-handler"]'

WebDriverWait(driver, 100).until(
    EC.presence_of_element_located((By.XPATH, Alle_ablehnen)))

# Find the Accept all element 
Accept = driver.find_element(By.XPATH , Alle_ablehnen )   
print(Accept.text)
actions.click(Accept)
actions.perform()


# Find the Job_title or Firma Search element     
Keyword_search = driver.find_element(By.XPATH , ' //*[@id="keywords"] ' ) 
print(Keyword_search.text)                            
actions.click(Keyword_search)
actions.perform()

Keyword = "python"
Keyword_search.send_keys(Keyword)

# Find the city Search  element 
city_search = driver.find_element(By.XPATH , '//*[@id="locations"]' )  
print(city_search.text)
actions.click(city_search)
actions.perform()

city = "wien"
city_search.send_keys(city)
city_search.send_keys(Keys.RETURN) 



for i in range(1,18):              

    try:
        job_cards = driver.find_element(By.XPATH , f'//*[@id="jobsearchListing"]/div[1]/div[1]/div[6]/ol/li[{i}]/div/div' )  
        actions.click(job_cards)                    
        actions.perform()   
    except NoSuchElementException:
        continue
                                                                     
    # Wait for the job_description element to appear on the page    
    WebDriverWait(driver, 100).until(
        EC.presence_of_element_located((By.XPATH, '/html/body/div[1]/div[3]/div/div[3]/div/div[2]/div/div[1]/div/div/div/div/div[4]')))

    job_description = driver.find_element(By.CLASS_NAME, 'm-jobContent__jobText' )     #m-jobContent  m-jobContent__jobDetail  m-jobContent__jobText
    print(job_description.text , '\n \n \n')

    with open('job_description_Karriere.txt', 'a') as f:
        f.write(job_description.text)
        f.write('\n \n \n' +'-'*160 + '\n \n \n')
```
