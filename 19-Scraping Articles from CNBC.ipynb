Scraping Articles from CNBC and Analyzing it .

```python
import ycnbc
from newspaper import Article
import pandas as pd

pd.set_option('display.max_colwidth', None)


data = ycnbc.News()

# get trending news
trending_ = data.trending() # return DataFrame

# get latest news
latest_ = data.latest() # return DataFrame

# get news by category
economy_ = data.economy() # return DataFrame


for link in economy_['Link'] :

    # Create an instance of the Article class
    article = Article(link)

    # Download and parse the article
    article.download()
    article.parse()

    print(article)

    article.nlp()

    # Access the summary and keywords
    print(article.summary , '\n')
    print(article.keywords, '\n' )

    # Print the article title, author, and text
    print("Title:", article.title)
    print("Author:", article.authors ,'\n \n')
    #print("Text:", article.text)



```
