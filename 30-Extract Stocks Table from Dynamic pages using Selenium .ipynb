Extract Stocks Table from Dynamic pages yFinance using Selenium.
Sector = Technology , Industry = Semiconductors

```python
from selenium import webdriver
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.support import expected_conditions as EC





# Create a new instance of the Chrome driver
driver = webdriver.Chrome(executable_path = 'chromedriver' )

# Navigate to the website you want to interact with
driver.get( "https://finance.yahoo.com/screener/equity/new")

WebDriverWait(driver, 10)

# Create an instance of ActionChains
actions = ActionChains(driver)

Accept_all = ' //*[@id="consent-page"]/div/div/div/form/div[2]/div[2]/button[1] '

# Find the Accept all element
element1 = driver.find_element(By.XPATH , Accept_all)   
print(element1.text)
actions.click(element1)
actions.perform()

# Find the Sector element
element2 = driver.find_element(By.XPATH , '//*[@id="screener-criteria"]/div[2]/div[1]/div[1]/div[4]/div/div[1]/div[2]/ul/li/div/div[1]/span/span ')
print(element2.text)
actions.click(element2)
actions.perform()

# choose the Sector element which Semiconductors
element3 = driver.find_element(By.XPATH , '//*[@id="dropdown-menu"]/div/div[2]/ul/li[11]/label/span ')
print(element3.text)                     # //*[@id="dropdown-menu"]/div/div[2]/ul/li[2]/label/span
actions.click(element3)                  # //*[@id="dropdown-menu"]/div/div[2]/ul/li[4]/label/span
actions.perform()


# close the window
element4 = driver.find_element(By.XPATH , '//*[@id="dropdown-menu"]/button')  
#print(element4.text)
actions.click(element4)
actions.perform()

WebDriverWait(driver, 10)            


# Wait for the Industry  element to appear on the page
element = WebDriverWait(driver, 10).until(
    EC.presence_of_element_located((By.XPATH, '//*[@id="screener-criteria"]/div[2]/div[1]/div[1]/div[4]/div/div[3]/div[2]/ul/li/div/div/span/span')))

#Find the Industry element
element5 = driver.find_element(By.XPATH , '//*[@id="screener-criteria"]/div[2]/div[1]/div[1]/div[4]/div/div[3]/div[2]/ul/li/div/div/span/span')
print(element5.text)                         
actions.click(element5)                   
actions.perform()

#choose the Industry element 
element6 = driver.find_element(By.XPATH , '//*[@id="dropdown-menu"]/div/div[2]/ul/li[11]/label/span')
print(element6.text)
actions.click(element6)
actions.perform()

#close the window 
element7 = driver.find_element(By.XPATH , '//*[@id="dropdown-menu"]/button')  
#print(element7.text)
actions.click(element7)
actions.perform()

#find Stocks
element8 = driver.find_element(By.XPATH , '//*[@id="screener-criteria"]/div[2]/div[1]/div[3]/button[1]/span')
print(element8.text)
actions.click(element8)
actions.perform()




# Close the browser
#driver.quit()
```

```python
from bs4 import BeautifulSoup

# Parse the HTML content of the page with BeautifulSoup
soup = BeautifulSoup(driver.page_source, 'lxml')

# Find the table of stock data
table = soup.find('table')

# Extract the column headers from the table
headers = [th.text.strip() for th in table.find_all('th')]

# Extract the rows of data from the table
rows = []
for tr in table.find_all('tr')[1:]:
    row = [td.text.strip() for td in tr.find_all('td')]
    rows.append(row)

# Print the column headers and the rows of data
print(headers)
for row in rows:
    print(row)

# Close the webdriver
#driver.quit()
```
