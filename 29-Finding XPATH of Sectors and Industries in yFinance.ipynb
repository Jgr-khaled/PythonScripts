Finding XPATH of Sectors and Industries in yFinance.      

```python
from selenium import webdriver
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoSuchElementException

import yaml

#option = webdriver.ChromeOptions()
#option.add_argument('--headless')
#options.add_argument('--no-sandbox')
#options.add_argument('--disable-dev-shm-usage')


# Create a new instance of the Chrome driver
driver = webdriver.Chrome(executable_path = 'chromedriver'  )

# Navigate to the website you want to interact with
driver.get( "https://finance.yahoo.com/screener/equity/new")

WebDriverWait(driver, 10)

# Create an instance of ActionChains
actions = ActionChains(driver)

Accept_all = ' //*[@id="consent-page"]/div/div/div/form/div[2]/div[2]/button[1] '

# Find the Accept all element
element1 = driver.find_element(By.XPATH , Accept_all)   
print(element1.text)
actions.click(element1)
actions.perform()


# Find the Sector element
element2 = driver.find_element(By.XPATH , '//*[@id="screener-criteria"]/div[2]/div[1]/div[1]/div[4]/div/div[1]/div[2]/ul/li/div/div[1]/span/span ')
print(element2.text)
actions.click(element2)
actions.perform()


sector_data = {}
for i in range(1,12):
   # choose the Sector element 
   element = driver.find_element(By.XPATH , f'//*[@id="dropdown-menu"]/div/div[2]/ul/li[{i}]/label/span ')
   print(element.text, f'    //*[@id="dropdown-menu"]/div/div[2]/ul/li[{i}]/label/span ')    
   sector_data[element.text] = f'//*[@id="dropdown-menu"]/div/div[2]/ul/li[{i}]/label/span '    

close = '//*[@id="dropdown-menu"]/button'
# close the window
element4 = driver.find_element(By.XPATH , close )  
actions.click(element4)
actions.perform()

#sectors = list(sector_data.items())
#print(sectors[0][1]) 

Industry = {}

for sector,sector_path  in sector_data.items():
                                               
   # Find the Sector element                   
   element2 = driver.find_element(By.XPATH , '//*[@id="screener-criteria"]/div[2]/div[1]/div[1]/div[4]/div/div[1]/div[2]/ul/li/div/div/span/span ')
   print(element2.text)                        
   actions.click(element2)                     
   actions.perform()                        

   # choose the Sector element 
   element3 = driver.find_element(By.XPATH , sector_path)
   print(element3.text)                  
   actions.click(element3)                  
   actions.perform()

   # close the window
   element4 = driver.find_element(By.XPATH , close )  
   actions.click(element4)
   actions.perform()

   WebDriverWait(driver, 10)            

   # Wait for the Industry  element to appear on the page
   WebDriverWait(driver, 10).until(
      EC.presence_of_element_located((By.XPATH, '//*[@id="screener-criteria"]/div[2]/div[1]/div[1]/div[4]/div/div[3]/div[2]/ul/li/div/div/span/span')))

   #Find the Industry element
   element5 = driver.find_element(By.XPATH , '//*[@id="screener-criteria"]/div[2]/div[1]/div[1]/div[4]/div/div[3]/div[2]/ul/li/div/div/span/span')
   actions.click(element5)                   
   actions.perform()

   Industry_data = {}
   for i in range(1,24):

      try:
         # choose the Sector element 
         elementt = driver.find_element(By.XPATH , f'//*[@id="dropdown-menu"]/div/div[2]/ul/li[{i}]/label/span ')
         print(elementt.text, f'   //*[@id="dropdown-menu"]/div/div[2]/ul/li[{i}]/label/span ')    
         Industry_data[elementt.text] = f' //*[@id="dropdown-menu"]/div/div[2]/ul/li[{i}]/label/span ' 
      except NoSuchElementException:
         break
   
   #//*[@id="dropdown-menu"]/div/div/ul/li[1]/label/span  Utilities  , Energy ,Communication Services
   if sector in ['Utilities', 'Energy', 'Communication Services']:
      for i in range(1,24):
         try:
            # choose the Sector element 
            elementtt = driver.find_element(By.XPATH , f'//*[@id="dropdown-menu"]/div/div/ul/li[{i}]/label/span ')
            print(elementtt.text, f'   //*[@id="dropdown-menu"]/div/div/ul/li[{i}]/label/span ')    
            Industry_data[elementtt.text] = f' //*[@id="dropdown-menu"]/div/div/ul/li[{i}]/label/span ' 
         except NoSuchElementException:
            break


   Industry[sector] = Industry_data
   
   #close the window 
   element7 = driver.find_element(By.XPATH , close)  
   actions.click(element7)
   actions.perform()
   
   #Remove the Sector 
   element7 = driver.find_element(By.XPATH , '//*[@id="screener-criteria"]/div[2]/div[1]/div[1]/div[4]/div/div[1]/div[2]/ul/li[1]/button/span')  
   actions.click(element7)
   actions.perform()

   WebDriverWait(driver, 10)


for sector,sector_path  in sector_data.items():
   print(Industry[sector])

with open('XPATH_Sectors_Industries.yaml', 'w') as f:
    yaml.dump(Industry, f)   

driver.quit()


```
