```python
from newscraping import news
from newscraping import newspapers
import pandas as pd

pd.set_option('display.max_rows', None)

available_sources = newspapers()
print(available_sources)
df = news(newspaper="reuters", n_articles=-1, early_date="2023-02-19", verbose=1)
df = df['headline']

headlines = df.to_string()


print(headlines)

```

```python

import nltk
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk import ne_chunk, pos_tag

# Tokenize the speech into words
words = word_tokenize(headlines)

# Tokenize the speech into sentences
sentences = sent_tokenize(headlines)
#print(sentences)



# Extract named entities
entities = ne_chunk(pos_tag(words), binary=False)
#print(entities)


```

```python
#Step 5: Analyze sentiment

from nltk.sentiment import SentimentIntensityAnalyzer

# Analyze sentiment
analyzer = SentimentIntensityAnalyzer()
sentiment = 0
for headline in df:
    sentiment += analyzer.polarity_scores(headline)['compound']
    
```

```python
#Step 6: Extract keywords

from nltk.corpus import stopwords
from string import punctuation
from collections import Counter

# Extract keywords
stop_words = set(stopwords.words('english') + list(punctuation))
keywords = [word for word in words if word.lower() not in stop_words]
#print(keywords)

word_countss = Counter(keywords)
top_words = word_countss.most_common(30)


print(f'Top 10 words: {top_words}')
print(f'Sentiment analysis: {sentiment}')
print(f'Keywords: {keywords}')
```
