```python
from selenium import webdriver
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver.common.keys import Keys

from bs4 import BeautifulSoup
import requests


# Create a new instance of the Chrome driver
driver = webdriver.Chrome(executable_path = 'chromedriver'  )

# Navigate to the website you want to interact with
driver.get("https://at.indeed.com/")       #"https://www.karriere.at/"

# Create an instance of ActionChains
actions = ActionChains(driver)

#WebDriverWait(driver, 20)

Alle_ablehnen  = ' //*[@id="onetrust-reject-all-handler"]'

# Find the Accept all element 
Accept = driver.find_element(By.XPATH , Alle_ablehnen )   
print(Accept.text)
actions.click(Accept)
actions.perform()


# Find the Job_title or Firma Search element     
Keyword_search = driver.find_element(By.XPATH , '//*[@id="text-input-what"]' )   
print(Keyword_search.text)
actions.click(Keyword_search)
actions.perform()

Keyword = "python"
Keyword_search.send_keys(Keyword)

# Find the city Search  element 
city_search = driver.find_element(By.XPATH , '//*[@id="text-input-where"]' )   
print(city_search.text)
actions.click(city_search)
actions.perform()

city = "wien"
city_search.send_keys(city)
city_search.send_keys(Keys.RETURN)


# close the google window  
google_anmelde = driver.find_element(By.XPATH , '//*[@id="google-Only-Modal"]/div/div[1]/button' )  
actions.click(google_anmelde)
actions.perform()

for i in range(1,18):
    
    try:
        job_cards = driver.find_element(By.XPATH , f'//*[@id="mosaic-provider-jobcards"]/ul/li[{i}]/div/div[1]/div/div[1]/div' )  
        actions.click(job_cards)                    
        actions.perform()   
    except NoSuchElementException:
        continue
    

    # Wait for the job_description element to appear on the page
    WebDriverWait(driver, 100).until(
        EC.presence_of_element_located((By.XPATH, '//*[@id="jobsearch-ViewjobPaneWrapper"]/div/div/div/div[2]/div/div/div[1]/div/div[1]')))

    job_description = driver.find_element(By.ID, 'jobDescriptionText' )  
    print(job_description.text , '\n \n \n')

    with open('job_description.txt', 'a') as f:
        f.write(job_description.text)
        f.write('\n \n \n' +'-'*160 + '\n \n \n')







```

```python
from selenium import webdriver
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver.common.keys import Keys

from bs4 import BeautifulSoup
import requests


# Create a new instance of the Chrome driver
driver = webdriver.Chrome(executable_path = 'chromedriver'  )

# Navigate to the website you want to interact with
driver.get("https://at.indeed.com/")       #"https://www.karriere.at/"

# Create an instance of ActionChains
actions = ActionChains(driver)

#WebDriverWait(driver, 20)

Alle_ablehnen  = ' //*[@id="onetrust-reject-all-handler"]'

# Find the Accept all element 
Accept = driver.find_element(By.XPATH , Alle_ablehnen )   
print(Accept.text)
actions.click(Accept)
actions.perform()


# Find the Job_title or Firma Search element     
Keyword_search = driver.find_element(By.XPATH , '//*[@id="text-input-what"]' )   
print(Keyword_search.text)
actions.click(Keyword_search)
actions.perform()

Keyword = "php and python"
Keyword_search.send_keys(Keyword)

# Find the city Search  element 
city_search = driver.find_element(By.XPATH , '//*[@id="text-input-where"]' )   
print(city_search.text)
actions.click(city_search)
actions.perform()

city = "wien"
city_search.send_keys(city)
city_search.send_keys(Keys.RETURN)


# close the google window  
google_anmelde = driver.find_element(By.XPATH , '//*[@id="google-Only-Modal"]/div/div[1]/button' )  
actions.click(google_anmelde)
actions.perform()






while True:

    j=2
    for i in range(1,18):
        
        try:
            job_cards = driver.find_element(By.XPATH , f'//*[@id="mosaic-provider-jobcards"]/ul/li[{i}]/div/div[1]/div/div[1]/div' )  
            actions.click(job_cards)                    
            actions.perform()   
        except NoSuchElementException:
            continue
    

        # Wait for the job_description element to appear on the page  
        WebDriverWait(driver, 100).until(
            EC.presence_of_element_located((By.XPATH, '//*[@id="jobsearch-ViewjobPaneWrapper"]/div/div/div/div[2]/div/div/div[1]/div/div[1]')))

        job_description = driver.find_element(By.ID, 'jobDescriptionText' )  
        print(job_description.text , '\n \n \n')

        with open('Jobs_description.txt', 'a') as f:
            f.write(job_description.text)
            f.write('\n \n \n' +'-'*160 + '\n \n \n')



    # Wait for the next_page element to appear on the page
    next_page = WebDriverWait(driver, 100).until(
        EC.presence_of_element_located((By.XPATH, f'//*[@id="jobsearch-JapanPage"]/div/div/div[5]/div[1]/nav/div[{j+1}]/a' )))
    
    if next_page :
        weiter = driver.find_element(By.XPATH , f'//*[@id="jobsearch-JapanPage"]/div/div/div[5]/div[1]/nav/div[{j}]/a' )  
        actions.click(weiter)                    
        actions.perform()
    else:
        break                                     
                                                
    j+=1




#//*[@id="jobsearch-JapanPage"]/div/div/div[5]/div[1]/nav/div[6]/a



#The code works perfectly but the website wants to prove that a human is visiting the website and nit a robot 


```
